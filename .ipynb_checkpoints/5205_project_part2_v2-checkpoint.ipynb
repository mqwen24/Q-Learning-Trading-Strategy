```python
import pandas as pd

def save_columns_to_csv(col_names, df, folder_name):
    # Create the folder if it doesn't exist
    os.makedirs(folder_name, exist_ok=True)
    
    # Iterate through each unique column name
    i=0
    for col in col_names:
        if col in df.columns:  # Ensure the column exists in the DataFrame
            df1 = df[col]
            if df1.isnull().all().all():
                print(f"All values are missing in column '{col}'. Skipping CSV creation.")
                continue
            df1.to_csv(f"{folder_name}/{col}.csv", index=True)
            print(f"{folder_name}/{col}.csv has been created.")
        else:
            print(f"Warning: Column '{col}' not found in DataFrame.")

save_columns_to_csv(unique_names, df_all, "Data/S&P500_All")
```

```python
import os
import pandas as pd

def load_csvs_to_dict(folder_name):
    data_dict = {}
    
    # Ensure the folder exists
    if not os.path.exists(folder_name):
        print(f"Error: Folder '{folder_name}' does not exist.")
        return data_dict

    # Iterate through the files in the folder
    for file in os.listdir(folder_name):
        if file.endswith(".csv"):  # Ensure only CSV files are processed
            file_path = os.path.join(folder_name, file)
            df_name = os.path.splitext(file)[0]  # Extract name without extension
            data_dict[df_name] = pd.read_csv(file_path, index_col=0)  # Read CSV with index
            
    return data_dict

# Test Code
folder = "Data/S&P500_All"
data_frames1 = load_csvs_to_dict(folder)

# Print the names of the DataFrames
print("Loaded DataFrames:", list(data_frames1.keys()))

# Display the first 5 rows and first 10 columns of each DataFrame
for name, df in data_frames1.items():
    print(f"\nDataFrame: {name}")
    display(df.iloc[:3, :7])  # Display first 5 rows and first 5 columns
```

```python
import pandas as pd

def combine_columns_from_df_dict(df_dict, n=0):
    """
    Combine the ith column from each DataFrame in df_dict into a new DataFrame.
    
    Parameters:
    - df_dict (dict of pd.DataFrame): Dictionary of dataframes with the same shape.
    - n (int): Number of columns to iterate over. If 0, use all columns.

    Returns:
    - result_dict (dict of pd.DataFrame): Dictionary with column names as keys and
      DataFrames with ith columns from all input DataFrames as values.
    """
    # Assume all dataframes have the same shape and columns
    first_df = next(iter(df_dict.values()))
    total_cols = len(first_df.columns)
    n = total_cols if n == 0 else min(n, total_cols)

    result_dict = {}

    for i in range(n):
        col_name = first_df.columns[i]
        combined_df = pd.DataFrame({
            key: df.iloc[:, i] for key, df in df_dict.items()
        })
        result_dict[col_name] = combined_df

    return result_dict

data_frames2 = combine_columns_from_df_dict(data_frames1)
```

```python
import pandas as pd
import numpy as np

# ---------------- Indicator Functions ----------------

def calculate_rsi(prices, window=14):
    """
    Calculate the Relative Strength Index (RSI) for a price series.
    """
    price_changes = prices.diff()
    gains = price_changes.where(price_changes > 0, 0)
    losses = abs(price_changes.where(price_changes < 0, 0))
    avg_gains = gains.rolling(window=window, min_periods=1).mean()
    avg_losses = losses.rolling(window=window, min_periods=1).mean()
    rs = avg_gains / avg_losses
    rsi = 100 - (100 / (1 + rs))
    return rsi

def calculate_macd(prices, short_window=12, long_window=26, signal_window=9):
    """
    Calculate the Moving Average Convergence Divergence (MACD) for a price series.
    """
    ema_short = prices.ewm(span=short_window, adjust=False).mean()
    ema_long = prices.ewm(span=long_window, adjust=False).mean()
    macd_line = ema_short - ema_long
    signal_line = macd_line.ewm(span=signal_window, adjust=False).mean()
    macd_df = pd.DataFrame({
        'macd_line': macd_line,
        'signal_line': signal_line
    })
    return macd_df

def calculate_stochastic_oscillator(df, window=14):
    # Placeholder — still does nothing, kept for consistency
    pass

# ---------------- Feature Engineering ----------------

def add_feature(df, n=10):
    df_ = df.copy()

    df_.index = pd.to_datetime(df_.index)

    # Basic range features
    df_["PCT_High_Close_Range"] = (df_["High"] - df_["Close"]) / df_["Close"]
    df_["PCT_Low_Close_Range"] = (df_["Close"] - df_["Low"]) / df_["Close"]
    df_["PCT_High_Low_Range"] = (df_["High"] - df_["Low"]) / df_["Close"]

    # Return & volatility
    df_["Return"] = df_["Close"].pct_change()
    df_["Volatility"] = df_["Return"].rolling(window=n).std() / (n ** 0.5)

    # Moving averages
    df_["MA15"] = df_["Close"].rolling(window=15).mean()
    df_["MA30"] = df_["Close"].rolling(window=30).mean()
    df_["MA150"] = df_["Close"].rolling(window=150).mean()

    # RSI
    df_["RSI"] = calculate_rsi(df_["Close"])

    # MACD
    macd_df = calculate_macd(df_["Close"])
    df_[["MACD", "Signal_MACD"]] = macd_df

    df_ = df_.drop(columns=["High", "Low", "Open"])

    return df_.dropna()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.pyplot as plt
import pandas as pd

def plot_columns(df, normalize=False):
    """
    Plot all columns in the DataFrame against the index (assumed to be Date).

    Parameters:
    - df (pd.DataFrame): DataFrame with DateTime index and numeric columns to plot.
    - normalize (bool): If True, normalize each column to [0, 1] range before plotting.
    """

    df_plot = df.copy()
    df_plot.index = pd.to_datetime(df_plot.index)

    # Normalize columns if requested
    if normalize:
        df_plot = (df_plot - df_plot.min()) / (df_plot.max() - df_plot.min())

    # Plot each column
    plt.figure(figsize=(12, 6))
    for column in df_plot.columns:
        plt.plot(df_plot.index, df_plot[column], label=column)

    plt.xlabel('Date')
    plt.ylabel('Normalized Value' if normalize else 'Value')
    title = 'Time Series Plot' + (' (Normalized)' if normalize else '')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(title, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

def sliding_window_forecast(df, model_class, window_size=14, model_params=None):
    if model_params is None:
        model_params = {}

    results = []

    df_plot = df.copy()
    df = (df - df.min()) / (df.max() - df.min())
    
    df = df_plot.reset_index()

    # Shift the target: predict next day's price using current features
    df["Target"] = df["Close"].shift(-1)

    # Drop the last row (Target is NaN)
    df = df.dropna()

    # Feature columns (exclude 'Date' and 'Target')
    feature_cols = [col for col in df.columns if col not in ['Date', 'Target']]

    for start in range(0, len(df) - window_size + 1):
        end = start + window_size
        window = df.iloc[start:end].copy()

        train_data = window.iloc[:-1]  # All except the last row
        test_data = window.iloc[-1:]   # Only the last row

        model = model_class(**model_params)
        model.fit(train_data[feature_cols], train_data["Target"])
        predictions = model.predict(test_data[feature_cols])

        temp_result = pd.DataFrame({
            "Date": test_data["Date"].values,
            "Actual": test_data["Target"].values,
            "Predicted": predictions
        })

        results.append(temp_result)

    final_results = pd.concat(results, ignore_index=True)
    return final_results
```

```python
import matplotlib.pyplot as plt

def plot_actual_vs_prediction(df, title="Actual vs Predicted Values"):
    """
    Plots the actual and predicted values over time.

    Parameters:
    - df: pandas DataFrame with columns ['Date', 'Actual', 'Predicted']
    - title: title of the plot (optional)
    """
    plt.figure(figsize=(14, 6))
    plt.plot(df['Date'], df['Actual'], label='Actual', linewidth=2)
    plt.plot(df['Date'], df['Predicted'], label='Predicted', linestyle='--', linewidth=2)
    
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Value")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
from typing import Dict, Type

def evaluate_models_limited(
    df: pd.DataFrame,
    model_dict: Dict[str, callable],
    window_size: int = 14
) -> Dict[str, pd.DataFrame]:
    """
    Evaluate multiple models on a DataFrame using sliding window forecast.

    Parameters:
    - df (pd.DataFrame): The feature-engineered DataFrame.
    - model_dict (dict): A dictionary of model names to model constructor functions (e.g., {"Linear": lambda: LinearRegression()}).
    - window_size (int): The number of rows in each sliding window.
    - train_ratio (float): Proportion of the window to use for training.

    Returns:
    - results (dict): A dictionary mapping model names to their prediction DataFrames.
    """
    results = {}
    for name, model_fn in model_dict.items():
        result_df = sliding_window_forecast(
            df,
            model_class=model_fn,
            window_size=window_size
        )
        results[name] = result_df
    return results

def summarize_r2_scores(results: Dict[str, pd.DataFrame]) -> pd.DataFrame:
    summary = {}

    for name, result_df in results.items():
        r2 = r2_score(result_df["Actual"], result_df["Predicted"])
        summary[name] = r2

    return pd.DataFrame.from_dict(summary, orient='index', columns=["Test R2"]).reset_index().rename(columns={"index": "Model"})
```
